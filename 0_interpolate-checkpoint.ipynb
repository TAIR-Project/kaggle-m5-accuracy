```python
datas = pd.read_csv(f"add_Clean_Analysis_{STORE}.csv").pipe(patair.reduce_mem_usage)
# dept_id,cat_id の情報を得るために、sales_train_validation.csvを読み込み
#  →マージして情報の取り込み
sales_train_validation = pd.read_csv('input/sales_train_validation.csv')
sales_train_validation = sales_train_validation.loc[:,['item_id','dept_id','cat_id']]
sales_train_validation
```

```python
datas=datas.merge(sales_train_validation,on='item_id')
datas = datas.iloc[:,1:]
datas
```

```python
def encode_categorical(df, cols):
    
    for col in cols:
        # Leave NaN as it is.
        le = LabelEncoder()
        #not_null = df[col][df[col].notnull()]
        df[col] = df[col].fillna('nan')
        df[col] = pd.Series(le.fit_transform(df[col]), index=df.index)

    return df
```

```python
datas = encode_categorical(
   datas, ["item_id",'dept_id','cat_id']
).pipe(patair.reduce_mem_usage)
gc.collect()
```

```python
datas.sort_values(["item_id", "variable", "week", "wday"], inplace=True)

datas["variable"] = datas["variable"].apply(lambda x: x[-2:])
datas["variable"] = datas["variable"].astype(np.int8)

# datas.drop("Unnamed: 0", axis=1, inplace=True)
datas = datas[datas["variable"] >= 13]
#順序の整った一意なキーを取得
product = pd.read_pickle(f"product/product_{STORE}.pickle")
```

```python
datas
```

```python
#一意なキーをnumpy配列で取得
columns = np.array(product["item_id"])
```

```python
# columns
```

```python
#各アイテムIDについて、0が15日以上続く区間で昨年の売り上げ(prev365)を射影
for l, item in enumerate(columns):
    buff = datas[datas["item_id"] == item]
    buff.reset_index(inplace=True)
    array = np.array(buff["value"])
    j = 0
    range_list = []
    for i in range(len(array)):
        if array[i] == 0:
            j += 1
        else:
            if j >= 15:
                range_list.extend(np.arange(i-j,i,1))
            j = 0
    if j >= 15:
        range_list.extend(np.arange(i-j+1,i+1,1))
    change_array = np.array(buff.loc[range_list]["index"])
    datas.loc[change_array,"value"] = datas.loc[change_array,"prev_1year"]
    if l % 300 == 0:
        print(f"{l}/3049")   
```

```python
datas.to_pickle(f"{STORE}_interpolate.pickle")
```
