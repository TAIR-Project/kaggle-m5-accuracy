```python
#売値(sell_price)を追加
df_term = pd.merge(df_term,df_sell_prices_CA_1,how = 'left',on = ["wm_yr_wk","item_id"])
df_term = df_term.drop(["store_id"],axis = 1)#店は固定で分析するので、store_idは不要
df_term
```

```python
#商品発売前のデータを削除
#( 売値(sell_prices)がNaNであるデータを削除 )
df_term = df_term.dropna(subset=["sell_price"])
df_term = df_term.reset_index(drop = True)
df_term
```

```python
#カテゴリー(cat_id)ごとの売り上げ個数の合計を求める
df_term_comp = df_term.drop(["wm_yr_wk","wday","month","year","sell_price"],axis = 1)
df_term_comp = df_term_comp.groupby(["cat_id","day"]).sum()
df_term_comp
```

```python
# FOODSだけのgroupを取得
df_term_comp_foods = df_term_comp.groupby('cat_id').get_group("FOODS")
df_term_comp_foods.reset_index(inplace=True)# reset_index
df_term_comp_foods
```

```python
# HOBBIESだけのgroupを取得
df_term_comp_hobbies = df_term_comp.groupby('cat_id').get_group("HOBBIES")
df_term_comp_hobbies.reset_index(inplace=True)# reset_index
df_term_comp_hobbies
```

```python
# HOUSEHOLDだけのgroupを取得
df_term_comp_household = df_term_comp.groupby('cat_id').get_group("HOUSEHOLD")
df_term_comp_household.reset_index(inplace=True)# reset_index
df_term_comp_household
```

```python
#HOUSEHOLDのグラフ
# import plotly.graph_objs as go
df_2012=df_term_comp_household[(df_term_comp_household['day']>=338 )&(df_term_comp_household['day']<=550)]
df_2012=df_2012.iloc[:,1:3]
df_2012['day'] = df_2012['day']-337
hh = df_2012

df_2013=df_term_comp_household[(df_term_comp_household['day']>=704 )&(df_term_comp_household['day']<=915)]
df_2013=df_2013.iloc[:,1:3]
df_2013['day'] = df_2013['day']-703
hh=hh.merge(df_2013,on='day',how='left')
hh=hh.rename(columns={'value_x':'2012','value_y':'2013'})

df_2014=df_term_comp_household[(df_term_comp_household['day']>=1068)&(df_term_comp_household['day']<=1280)]
df_2014=df_2014.iloc[:,1:3]
df_2014['day'] = df_2014['day']-1068
hh=hh.merge(df_2014,on='day',how='left')

df_2015=df_term_comp_household[(df_term_comp_household['day']>=1433)&(df_term_comp_household['day']<=1645)]
df_2015=df_2015.iloc[:,1:3]
df_2015['day'] = df_2015['day']-1433
hh=hh.merge(df_2015,on='day',how='left')

df_2016=df_term_comp_household[(df_term_comp_household['day']>=1799)&(df_term_comp_household['day']<2000)]
df_2016=df_2016.iloc[:,1:3]
df_2016['day'] = df_2016['day']-1798
hh=hh.merge(df_2016,on='day',how='left')
# hh=hh.drop('day',axis=1)
hh=hh.rename(columns={'value_x':'2014','value_y':'2015','value':'2016'})
hh
```
