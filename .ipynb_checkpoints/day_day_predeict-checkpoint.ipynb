```python
def encode_categorical(df, cols):
    
    for col in cols:
        # Leave NaN as it is.
        le = LabelEncoder()
        #not_null = df[col][df[col].notnull()]
        df[col] = df[col].fillna('nan')
        df[col] = pd.Series(le.fit_transform(df[col]), index=df.index)

    return df
```

```python
#最大ラグ
max_lag = 37
test_day = 28

#時系列特徴
time_features = ['prev_1','prev_2','prev_7', 'prev_14', 'prev_28','p_prev_1', 'rolling_mean_t7', 'rolling_mean_t30', 
                 'rolling_std_t7', 'rolling_std_t30', 'price_change_t1', 'rolling_price_std_t7',
                 'rolling_price_std_t30']
```

```python
datas = pd.read_pickle("dataset_CA_contain_null.pickle")
```

```python
datas
```

```python
datas.drop(time_features, axis=1, inplace=True)
```

```python
result = datas[datas["date"] == datetime(2016,4,25)][["id"]]
```

```python
for i in range(test_day):
    buff = datas[(datas["date"] >= (datetime(2016,4, 25) + timedelta(days=-max_lag + i))) &  (datas["date"] <= datetime(2016,4, 25) + timedelta(days=i))].copy()
    
    #1日前の売り上げ個数を特徴に追加
    buff.loc[buff["date"] ==  datetime(2016,4, 25) + timedelta(days=i), "prev_1"] = \
    np.array(pd.merge(left=result ,right=buff.groupby("id").nth(-2)[["sell_num"]].reset_index(), how="left")["sell_num"])
    
    #2日前の売り上げ個数を特徴に追加
    buff.loc[buff["date"] ==  datetime(2016,4, 25) + timedelta(days=i), "prev_2"] = \
    np.array(pd.merge(left=result ,right=buff.groupby("id").nth(-3)[["sell_num"]].reset_index(), how="left")["sell_num"])
    
    #7日前の売り上げ個数を特徴に追加
    buff.loc[buff["date"] ==  datetime(2016,4, 25) + timedelta(days=i), "prev_7"] = \
    np.array(pd.merge(left=result ,right=buff.groupby("id").nth(-8)[["sell_num"]].reset_index(), how="left")["sell_num"])

    #14日前の売り上げ個数を特徴に追加
    buff.loc[buff["date"] ==  datetime(2016,4, 25) + timedelta(days=i), "prev_14"] = \
    np.array(pd.merge(left=result ,right=buff.groupby("id").nth(-15)[["sell_num"]].reset_index(), how="left")["sell_num"])

    #28日前の売り上げ個数を特徴に追加
    buff.loc[buff["date"] ==  datetime(2016,4, 25) + timedelta(days=i), "prev_28"] = \
    np.array(pd.merge(left=result ,right=buff.groupby("id").nth(-29)[["sell_num"]].reset_index(), how="left")["sell_num"])


    #1日前の価格を特徴に追加
    buff.loc[buff["date"] ==  datetime(2016,4, 25) + timedelta(days=i), "p_prev_1"] = \
    np.array(pd.merge(left=result ,right=buff.groupby("id").nth(-2)[["sell_price"]].reset_index(), how="left")["sell_price"])

    #窓平均
   # for size in [7, 30]:
   #     #自分とその前数日の平均    
   #     buff.loc[buff["date"] ==  datetime(2016,4, 25) + timedelta(days=i), f"rolling_mean_t{size}"] = \
   #     np.array(buff.groupby("id").tail(7+size).groupby("id")["sell_num"].transform(
   #         lambda x: x.shift(7).rolling(size).mean()
   #     ).dropna())
    
    for size in [7, 30]:
        #自分とその前数日の平均    
        buff.loc[buff["date"] ==  datetime(2016,4, 25) + timedelta(days=i), f"rolling_mean_t{size}"] = \
        np.array(buff.groupby("id").tail(size+1).groupby("id")["sell_num"].transform(
            lambda x: x.shift(1).rolling(size).mean()
    ).dropna())


    #窓偏差
    for size in [7, 30]:    
        buff.loc[buff["date"] ==  datetime(2016,4, 25) + timedelta(days=i), f"rolling_std_t{size}"] = np.array(buff.groupby("id").tail(size+1).groupby("id")["sell_num"].transform(
        lambda x: x.shift(1).rolling(size).std()
     ).dropna())

    #値段変化
    buff.loc[buff["date"] ==  datetime(2016,4, 25) + timedelta(days=i), "price_change_t1"] = \
    np.array(pd.merge(left=result ,right=pd.DataFrame((buff.groupby("id").nth(-1)["p_prev_1"] - buff.groupby("id").nth(-1)["sell_price"]) / buff.groupby("id").nth(-1)["p_prev_1"], columns=["a"]).reset_index(), how="left")["a"])

    buff.loc[buff["date"] ==  datetime(2016,4, 25) + timedelta(days=i), "rolling_price_std_t7"] = np.array(buff.groupby("id").tail(7).groupby("id")["sell_price"].transform(
        lambda x: x.rolling(7).std()
    ).dropna())
    buff.loc[buff["date"] ==  datetime(2016,4, 25) + timedelta(days=i), "rolling_price_std_t30"] = np.array(buff.groupby("id").tail(30).groupby("id")["sell_price"].transform(
        lambda x: x.rolling(30).std()
    ).dropna())
    
    buff = buff[buff["date"] == datetime(2016,4, 25) + timedelta(days=i)]
    datas.loc[datas["date"] == datetime(2016,4, 25) + timedelta(days=i),"sell_num"] = model.predict(buff[features])
    
    
    result[f"F{i+1}"] = model.predict(buff[features])
    print(f"{i+1}/{test_day}")
```

```python
result.to_pickle("result_f4.pickle")
datas.to_pickle("result_f_features4.pickle")
```
